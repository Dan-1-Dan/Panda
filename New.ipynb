```python
import pandas as pd
sda = {'Name': ['Dan', 'Lisa', 'Tom'], 'Age': [25, 30, 22]}
df = pd.DataFrame(sda)
print(df)
df[df['Age'] > 25]
#csv = pd.read_csv('data.csv')   
#excel = pd.read_excel('data.xlsx')
df["Age_Group"] = df["Age"].apply(lambda x: "Adult" if x >= 25 else "Young")

older_than_25 = df[df["Age"] > 25]
print("\nPeople older than 25:")
print(older_than_25)
young_people = df[df["Age_Group"] == "Young"]
print(young_people)

age_summary = df["Age"].agg(["mean", "max", "min"])


print("\nAge summary:")
print(age_summary)
```
